# T1090 Proxy

-----------------------------------------------------------------------

## Technique Description

Adversaries may use a connection proxy to direct network traffic between systems or act as an intermediary for network communications to a command and control server to avoid direct connections to their infrastructure. Many tools exist that enable traffic redirection through proxies or port redirection, including [HTRAN](https://attack.mitre.org/software/S0040), ZXProxy, and ZXPortMap. (Citation: Trend Micro APT Attack Tools) Adversaries use these types of proxies to manage command and control communications, reduce the number of simultaneous outbound network connections, provide resiliency in the face of connection loss, or to ride over existing trusted communications paths between victims to avoid suspicion. Adversaries may chain together multiple proxies to further disguise the source of malicious traffic.

Adversaries can also take advantage of routing schemes in Content Delivery Networks (CDNs) to proxy command and control traffic.

## Technique Detection

Analyze network data for uncommon data flows (e.g., a client sending significantly more data than it receives from a server or between clients that should not or often do not communicate with one another). Processes utilizing the network that do not normally have network communication or have never been seen before are suspicious. Analyze packet contents to detect communications that do not follow the expected protocol behavior for the port that is being used. (Citation: University of Birmingham C2)

Consider monitoring for traffic to known anonymity networks (such as [Tor](https://attack.mitre.org/software/S0183)).

## Host Analytics

-----------------------------------------------------------------------

### Hunter Notes

- If UPnP is enabled on the network, then there may be a large amount of traffic. Detection will be more difficult.
- UPnP traffic is not seen often and may appear on different ports
- There is a possibility that not all traffic will be routed through another host, but instead only some or certain types of the traffic will be proxied through another host on the network. This will make detection more difficult.

## Network Analytics

-----------------------------------------------------------------------

### Hunter Notes

- If UPnP is enabled on the network, then there may be a large amount of traffic. Detection will be more difficult.

- UPnP traffic is not seen often and may appear on different ports

- There is a possibility that not all traffic will be routed through another host, but instead only some or certain types of the traffic will be proxied through another host on the network. This will make detection more difficult.

- Review of router configurations and NAT tables will assist in identifying any modifications that are not standard to the MP network.

#### Analytic 1

  * **Information:** dentifying possible RDP connections coming into the network from external sources. If RDP server is located pivot to who it is communicating with. (5 W's) 

  * **Source:** Network Traffic, PCAP

  * **Tool:** Arkime, Kibana

  * **Notes:** If an external source utilizing RDP is identified, pivot to the external IP and document additional activities

  * **Arkime Query:** ```protocols == [rdp, dcerpc] && ip.src != [10/8, 192.168/12, 172.16/12]```
  
  * **Kibana Query:** ```protocol: (rdp or dcerpc) AND NOT scrIp: (10/8, 192.168/12, 172.16/12)```

#### Analytic 2

  * **Information:** Identify internal host to internal host routing and what ports or services they may be using. Correlate these services with the MP.

  * **Source:** Network Traffic, PCAP

  * **Tool:** Arkime, Kibana

  * **Notes:** Query will need to be modified to filter out normal/benign traffic to scope

  * **Arkime Query:** `ip.src == <internal host> && ip.dst == [10/8, 192.168/12, 172.16/12]`
  
  * **Kibana Query:** `srcIp: <internal host> AND dstIp: (10/8, 192.168/12, 172.16/12)`
  
#### Analytic 3

  * **Information:** Identify hosts that are broadcasting themselves to the network

  * **Source:** Network Traffic, PCAP

  * **Tool:** Arkime, Kibana

  * **Notes:** MP request will need to be submitted to identify if SSDP/uPnP is allowed and able to be utilized

  * **Arkime Query:** `port == 1900 || protocols == ssdp`
