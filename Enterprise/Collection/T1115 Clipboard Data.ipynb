# T1115 Clipboard Data

-----------------------------------------------------------------------

## Technique Description

Adversaries may collect data stored in the clipboard from users copying information within or between applications. 

In Windows, Applications can access clipboard data by using the Windows API.(Citation: MSDN Clipboard) OSX provides a native command, <code>pbpaste</code>, to grab clipboard contents.(Citation: Operating with EmPyre)

## Technique Detection

Access to the clipboard is a legitimate function of many applications on an operating system. If an organization chooses to monitor for this behavior, then the data will likely need to be correlated against other suspicious or non-user-driven activity.

#### APT28	
- A JHUHUGIT variant (malware used by APT28) accesses a screenshot saved in the clipboard and converts it to a JPG image.
- Kodiac (post-exploitation framework used by APT28) can retrieve the current content of the user clipboard.

#### APT29	
- CosmicDuke (malware used by APT29) copies and exfiltrates the clipboard contents every 30 seconds.

#### Turla	
- Empire (post-exploitation framework used by Turla) can harvest clipboard data on both Windows and macOS systems.

-----------------------------------------------------------------------

### Hunter Notes

- Access to the clipboard is a legitimate function of many applications on an operating system.
- If an organization chooses to monitor for this behavior, then the data will likely need to be correlated against other suspicious or non-user-driven activity.

#### Analytic 1

  * **Information:** PS Module for clipboard data

  * **Source:**  Windows Audits, Sysmon

  * **Tool:** Kibana

  * **Notes:** 

  * **Query:** ```event.code : 4104 AND message contains Get-Clipboard or Set-clipboard```

#### Analytic 2

  * **Information:** With the release of Sysmon 12, users can now configure the utility to generate an event whenever data is copied to the Clipboard. Clipboard data is also saved in files that are only accessible to an administrator for later review.

  * **Source:**  Windows Audits, Sysmon

  * **Tool:** Kibana

  * **Notes:** 

  * **Query:** ```event.code : 24```

#### Analytic 3

  * **Information:** clip.exe = Copy the result of any command to the Windows clipboard.

  * **Source:**  Windows Audits, Sysmon

  * **Tool:** Kibana

  * **Notes:** 

  * **Query:** ```process.name : clip.exe```
